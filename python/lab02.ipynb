```python
#logical statements and/or/not

num1 = 10
num2 = 20
print(num1 >= 5 and num2 <= 40) # and not &&
print(num1 >= 5 and num2 <= 19) # false bc num 2 is less than 20
print()

colour = "blue"
print(colour == "blue" or colour == "red" or colour == "green")

print("u" in colour or len(colour) == 3) # len gets the number of items in an object
print()

# not is when you want to reverse the result
mark = 70
print(not mark == 70) # false with not but delete not it is true
```

```python
# if statements
myBool = True
if myBool:
  print("success")

a = 33
b = 200
if b > a:
  print("b is greater than a")

x = 10
y = 20
if x < y:
  print("x is less than y") # this will print
else:
  print("y is greater than x")

print()
# multi way if statements (else if)
i = 4
if i == 1:
  print("1")
elif i == 2:
  print("2")
elif i == 3:
  print("3")
elif i == 4:
  print("4") # this will print
print()
# nested if

hourlyEmployee = True
hoursWorked = 30
bonus = 0

if hourlyEmployee == True:
  if hoursWorked > 40:
    bonus = 500
  else:
    bonus = 100
else:
  if hoursWorked < 20:
    bonus = 0
  else:
    bonus = 100

print("hourly employee: ", hourlyEmployee, "\n hours worked: ", hoursWorked,
      "\n bonus: ", bonus)

print()

# terniary

examScore = 100
grade = "A+" if examScore >= 95 else "B" # prints A+ since  grade is greater than 95
print(grade)
```
