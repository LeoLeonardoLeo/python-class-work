```python
#strings
str1 = "a double quoted string"
str2 = 'a single quoted string'
print(str1)
print(str2)

print()

#float
print(10.55)

flt1 = 100.434
print(flt1)

print()

#boolean (must be capital)
print(True)
print(False)

bool_t = True
bool_f = False

print(bool_t)
print(bool_f)

print()

#numeric operations
num1 = 123
num2 = 456
result = num1 + num2
print(result)

result2 = num1 + num2 * num2 - num1 ** num1 / num1
print(result2)

print()

#order of operations

a = 10
b = -6
c = 20
d = 45
total1 = (a - b) * c / d
total2 = ((c * c) + a) - d
print(total1)
print(total2)

print()

#concatenation

string_1 = "hello"
string_2 = "world"
print(string_1 + " " + string_2)

print()

#splitting statements

print(
    "hello " +
    "there " +
    "from " +
    "multiple " +
    "lines ")

print()

#common functions

print("hello", 123, True, 54.55)

print()

#user input

name = input("Enter your name: ")

print("Hello " + name)
```

```python
#compound

num1 = 10
num2 = 20

num1 += 7;
print(num1)

num2 *= 4
print(num2)
```
